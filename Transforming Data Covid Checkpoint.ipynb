```python
import pandas as pd 
```

```python
df = pd.read_csv("covid_tracking.csv")

df
```

```python
# Rename the column
df.rename(columns={'dataRep': 'Date'}, inplace=True)
```

```python
# Duplicate the Date column
df['Date_duplicated'] = df['date']
```

```python
# Split the Date column into Day, Month, and Year
df[['Day', 'Month', 'Year']] = df['date'].str.split('/', expand=True)
```

```python
# Convert Day, Month, and Year to appropriate data types
df['day'] = pd.to_numeric(df['day'], errors='coerce')
df[';onth'] = pd.to_numeric(df['month'], errors='coerce')
df['year'] = pd.to_numeric(df['year'], errors='coerce')



```

```python
# Replace NaN values with 0
df.fillna(0, inplace=True)
```

```python
# Rename columns with more meaningful names
df.rename(columns={
    'Day': 'Day_of_Month',
    'Month': 'Month_of_Year',
    'Year': 'Year_of_Observation',
    'date': 'Original_Date',
    'Date_duplicated': 'Duplicated_Date'
}, inplace=True)

df
```
